## Thinking about Vector Coordinates
- Each coordinate as a scalar that stretches or squishes vectors.
- Special vectors in the xy-coordinate system: i-hat (unit vector in x-direction) and j-hat (unit vector in y-direction).
- Framing coordinates as scalars that scale basis vectors.

## Basis Vectors
- Basis vectors: i-hat and j-hat in standard coordinate systems.
- Any pair of linearly independent vectors can form a basis for a coordinate system.
- Choices of basis vectors lead to different coordinate systems.

## Linear Combinations and Span
- Linear combination: Adding together scaled vectors.
- Linear combinations of i-hat and j-hat result in a span covering all 2D space.
- Span of vectors: Set of all possible vectors achievable through linear combinations.
- The concept of span becomes more interesting in three-dimensional space.

## Span in 3D Space
- Span of two vectors in 3D space: Forms a flat sheet.
- Adding a third vector: Can either sit on the span or unlock access to all 3D space.
- Linearly dependent vectors: One is redundant, and removing it doesn't reduce the span.
- Linearly independent vectors: Each adds a new dimension to the span.

## Basis of a Space
- Technical definition: A basis is a set of linearly independent vectors that span a space.
- Puzzle: Why does this definition make sense? Think about the relationships between span and linear independence.
