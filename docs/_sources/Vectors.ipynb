## Linear Algebra and Vectors
- "The introduction of numbers as coordinates is an act of violence." —Hermann Weyl
- Fundamental building block for linear algebra: the vector

## Perspectives on Vectors
1. **Physics Student Perspective:**
   - Vectors are arrows in space.
   - Defined by length and direction.
   - Can be moved around while preserving these characteristics.
   - Two-dimensional vectors in a flat plane, three-dimensional in broader space.

2. **Computer Science Student Perspective:**
   - Vectors are ordered lists of numbers.
   - Example: House features represented as a pair of numbers (e.g., square footage and price).
   - Order matters; vectors are modeled as two-dimensional lists.

3. **Mathematician's Perspective:**
   - Generalizes both physics and computer science views.
   - Vector is anything with sensible addition and scalar multiplication operations.
   - Abstract concept, focus on operations.

## Coordinate System
- Emphasis on a geometric perspective.
- Two dimensions: x-axis and y-axis.
- Origin at their intersection.
- Coordinates represent instructions from the origin to the vector's tip.

## Three Dimensions
- Add a z-axis perpendicular to x- and y-axes.
- Vectors associated with ordered triplets of numbers.
- Each triplet uniquely corresponds to one vector in space.

## Vector Operations: Addition
- Add two vectors by placing the tail of the second at the tip of the first.
- The sum is the vector from the first's tail to the second's tip.
- Vector addition mirrors combining movements in space.

### Numerical Example
- Vectors: (1, 2) and (3, -1).
- Tip-to-tail addition: "walk 1 to the right, then 2 up, then 3 to the right, then 1 down."
- Numerical addition: (1 + 3, 2 + (-1)).

## Vector Operations: Scalar Multiplication
- Multiply a vector by a number (scalar).
- Scaling: Stretch, squish, or reverse the direction of a vector.
- Numerically: Multiply each component by the scalar.

## Fundamental Operations
- Linear algebra centers around vector addition and scalar multiplication.
- Mathematician's viewpoint: Abstract operations.

## Conclusion
- Linear algebra's power lies in translating between geometric and numerical perspectives.
- Useful for data analysis, providing visual conceptualization for lists of numbers.
- Essential for physicists and computer graphics programmers to describe space and manipulations numerically.
