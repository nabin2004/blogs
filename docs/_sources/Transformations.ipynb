# Linear Algebra: Matrix-Vector Multiplication and Transformations


#### Introduction
- Linear algebra's key concept: **Linear Transformation** and its relation to matrices.
- Emphasis on understanding visually and avoiding memorization.

#### Linear Transformation Defined
- **Transformation = Function:** Takes in vectors, produces outputs.
- Visualize vectors as points; transformations as movements in space.
- Simplify by representing vectors as points for clarity.

#### Linear vs. Non-linear Transformations
- **Linear Transformations:**
  - Lines remain lines.
  - Origin stays fixed.
- Visual examples of linear and non-linear transformations.

#### Matrix-Vector Multiplication
- Describing linear transformations numerically.
- Basis vectors, i-hat and j-hat, land at specific coordinates.
- Any vector can be deduced using a linear combination of basis vectors.

#### Matrix Representation
- Two-dimensional linear transformations described by a **2x2 matrix**.
- Columns represent transformed basis vectors.
- Matrix interpretation simplifies understanding transformations.

#### Practice with Matrices
1. **Rotation Matrix:**
   - 90° counterclockwise rotation.
   - Matrix columns: (0, 1), (-1, 0).

2. **Shear Transformation:**
   - i-hat fixed, j-hat sheared to (1, 1).
   - Matrix columns: (1, 0), (1, 1).

3. **Deducing Transformation from Matrix:**
   - Start with columns (1, 2) and (3, 1).
   - Move i-hat and j-hat to respective coordinates.

#### Linearly Dependent Vectors
- Dependent vectors lead to squishing space onto a line.
- Linear transformations are one-dimensional in such cases.

#### Key Takeaways
- **Linear Transformations:** Grid lines remain parallel, origin fixed.
- **Matrix Representation:** Columns define transformed basis vectors.
- **Matrix-Vector Multiplication:** Compute transformations numerically.
- **Visualization:** Understand transformations visually, interpret matrices as space transformations.
